```python
from google_play_scraper import app, reviews, Sort
import pandas as pd
```

```python
bca_info = app('com.bca')
print(bca_info)
```

```python
bca_reviews, _ = reviews(
    'com.bca',
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=12000
)
```

```python
df_reviews = pd.DataFrame(bca_reviews)
df_reviews.head()
```

```python
df_reviews.shape
```

```python
df_reviews.to_csv('bca_mobile_reviews.csv', index=False)
```
